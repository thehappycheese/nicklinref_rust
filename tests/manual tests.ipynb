```python
import aiohttp
from urllib.parse import urlencode
import json
```

# GET Point

```python
async with aiohttp.ClientSession() as session:
        print("OLD ENDPOINT")
        async with session.get(
            'http://127.0.0.1:8080/?road=H001&slk=10',
        ) as resp:
            print(resp.status)
            print(await resp.text())
        print("NEW ENDPOINT")
        async with session.get(
            'http://127.0.0.1:8080/point?road=H001&slk=10',
        ) as resp:
            print(resp.status)
            print(await resp.text())
```

# POST Point

```python
async with aiohttp.ClientSession() as session:
    async with session.post(
            'http://127.0.0.1:8080/point',
            json={
                "road":"H001",
                "slk":10
            }
        ) as resp:
        print(resp.status)
        print(await resp.text())
            
```

# GET Line

```python
async with aiohttp.ClientSession() as session:
        async with session.get(
            'http://127.0.0.1:8080/?road=H001&slk_from=10&slk_to=20',
        ) as resp:
            print(resp.status)
            print(await resp.text())
        print("NEW LINE API") 
        async with session.get(
            'http://127.0.0.1:8080/?road=H001&slk_from=10&slk_to=20',
        ) as resp:
            print(resp.status)
            print(await resp.text())
```

# POST Line

```python
async with aiohttp.ClientSession() as session:
        async with session.post(
            'http://127.0.0.1:8080/line',
            json={
                    "road":"H001",
                    "slk_from":10,
                    "slk_to":20,
                    "offset":10
              }
        ) as resp:
            print(resp.status)
            print(await resp.text())
```

# POST Batch2

```python
query_parameters = {
    "format":"wkt",
    "items":[
        {
            "road":"H001",
            "slk_from":10,
            "slk_to":20,
            "offset":10
        },
        {
            "road":"H016",
            "slk":10
        },
        {
            "road":"H015",
            "slk":10
        }
    ]
}
```

```python
async with aiohttp.ClientSession() as session:
    async with session.post(
        'http://127.0.0.1:8080/batch2/',
        json=query_parameters
    ) as resp:
        print(resp.status)
        print(await resp.text())
```

# GET batch2

```python
async with aiohttp.ClientSession() as session: 
    async with session.get(
        f'http://127.0.0.1:8080/batch2/?{urlencode({**query_parameters, "items":json.dumps(query_parameters["items"])})}'
    ) as resp:
        print(resp.status)
        print(await resp.text())
```
